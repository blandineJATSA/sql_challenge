pour analyser les performances des pages d'atterrissage et comparer plusieurs pages, nous allons encore une fois utiliser des tables temporaires et écrire un programme de données à plusieurs étapes.

Nous allons trouver la première vue de page pour les sessions pertinentes. C'est ainsi que nous identifions notre page d'atterrissage.

Ensuite, nous associerons cette vue de page avec l'URL qu'ils ont vue.

Ensuite, nous analyserons si cette session a eu d'autres vues de pages. C'est ainsi que nous déterminons si nous qualifierons cette session de rebond.

Les rebonds n'ont aucune vue de page après cette première vue de page, tandis que les sessions non rebondies ont d'autres vues de pages après la première vue de page.

```python
query = """ 


WITH first_pageview AS (
SELECT session_id,
       MIN(id) min_pv_id              
FROM events
GROUP BY session_id
),

landing_page_demo AS (
SELECT first_pageview.session_id,
       events.event_type AS landing_page,
FROM first_pageview
LEFT JOIN events ON first_pageview.min_pv_id = events.id 
),
bounced_sessions_only AS (
SELECT landing_page_demo.session_id,
       landing_page_demo.landing_page,
       COUNT(DISTINCT events.id) AS count_of_pages_viewed
FROM landing_page_demo
LEFT JOIN events ON landing_page_demo.session_id =events.session_id
GROUP BY landing_page_demo.session_id,
         landing_page_demo.landing_page
HAVING count_of_pages_viewed=1
)
SELECT landing_page_demo.landing_page,
       COUNT(DISTINCT landing_page_demo.session_id) AS sessions,
       COUNT(DISTINCT bounced_sessions_only.session_id) AS bounced_sessions,
       ROUND(COUNT(DISTINCT bounced_sessions_only.session_id)/COUNT(DISTINCT landing_page_demo.session_id),3) AS bounce_rate
FROM landing_page_demo 
LEFT JOIN bounced_sessions_only ON landing_page_demo.session_id = bounced_sessions_only.session_id
GROUP BY landing_page_demo.landing_page
ORDER BY bounce_rate DESC
"""

db.sql(query)
```

on n'a notre table landing page pour chaque session_id
2ieme etape, pour chaque session, on compte le nombre de page et on ne selectionne que les sessions dont le nombre de page est 1 pour etre sur qu'après cette page, on obtient le resultat dans la table bounced_session_only
3ieme etape, on affiche toute les landing page puis les sessions de tout les landing pages y compris les landings pages qui sont consideres comme etant les seuls pages de la sessions et on fait une jointure à gauche avec la table des sessions (table bounced_session_only) de l'etape precedente qui ne contient que des sessions ayant une seule page et donc les sessions qui ont ont plusieurs pages auront des nulls lors de la jointure
