## NTILE

### NTILE avec ORDER BY

```python
inventory_items.head(2)
```

```python
# Écrire une requête qui divise les produits en 4 quartiles en fonction de leur prix de détail (product_retail_price) 

query = """ 
SELECT id,
       product_name,
       NTILE(4) OVER(ORDER BY product_retail_price) AS product_quartile
FROM inventory_items
LIMIT 1000 

"""

db.sql(query)
```

### NTILE avec PARTITION BY

```python
# Écrire une requête qui divise les produits en 3 tiers (terciles) pour chaque catégorie de produit (product_category) en fonction de 
#leur prix de détail (product_retail_price) et affiche l'identifiant, le nom du produit, la catégorie et le tercile auquel
# appartient chaque produit.

query = """ 
SELECT 
    id,
    product_name,
    product_category,
    product_retail_price,
    NTILE(3) OVER (PARTITION BY product_category ORDER BY product_retail_price) AS category_price_tercile
FROM 
    inventory_items
"""

db.sql(query)
```

### NTILE avec ORDER BY et PARTITION BY 

```python

# Écrire une requête qui divise les utilisateurs en 5 quintiles pour chaque état (state) en fonction de leur âge, affiche l'identifiant, 
#le nom, l'état, l'âge et le quintile auquel appartient chaque utilisateur, et trie les résultats par état et quintile.

query = """ 
SELECT 
    id,
    first_name,
    last_name,
    state,
    age,
    NTILE(5) OVER (PARTITION BY state ORDER BY age) AS age_quintile
FROM 
    users
ORDER BY 
    state, age_quintile;


"""

db.sql(query)
```

## LAG

### LAG avec ORDER BY 

```python
# Écrire une requête le prix actuel et le prix du produit précédent (selon l'ordre de prix) pour chaque produit.

query = """  
SELECT id,
       product_name,
       product_retail_price,
       LAG(product_retail_price) OVER(ORDER BY product_retail_price) AS previous_price
FROM inventory_items


"""

db.sql(query)
```

```python
# Calcul des différences de prix entre deux produits consécutifs (Récupère le prix de détail du produit précédent selon l'ordre des prix et 
# Calcule la différence de prix entre le produit actuel et le produit précédent.)

query = """  
SELECT 
    id,
    product_name,
    product_retail_price,
    LAG(product_retail_price) OVER (ORDER BY product_retail_price) AS previous_price,
    product_retail_price - LAG(product_retail_price) OVER (ORDER BY product_retail_price) AS price_difference
FROM 
    inventory_items;


"""
db.sql(query)
```

### LAG avec ORDER BY et PARTITION BY 

```python
# Suivre les changements d'état d'une commande(voir comment l'état d'une commande change d'une étape à l'autre.)

query = """  

SELECT 
    order_id,
    status,
    created_at,
    LAG(status) OVER (PARTITION BY order_id ORDER BY CAST(created_at AS TIMESTAMP)) AS previous_status
FROM 
    order_items;

"""

db.sql(query)
```

## FIRST_VALUE

```python
# Obtenir le produit le moins coûteux dans chaque catégorie 

query = """  
SELECT 
    category,
    name,
    retail_price,
    FIRST_VALUE(name) OVER (PARTITION BY category ORDER BY retail_price) AS cheapest_product_in_category
FROM 
    products;
"""

db.sql(query)
```

## QUALIFY

```python
#  Obtenir les utilisateurs les plus récents par ville
# Cette requête récupère les utilisateurs les plus récemment inscrits pour chaque ville.

query = """  
SELECT 
    city,
    id,
    first_name,
    last_name,
    created_at,
    ROW_NUMBER() OVER (PARTITION BY city ORDER BY created_at DESC) AS user_rank
FROM 
    users
QUALIFY user_rank = 1

"""

db.sql(query)
```

```python
# Obtenir les articles en stock les plus anciens pour chaque produit
# Cette requête récupère l'article en stock le plus ancien pour chaque produit.

query = """  
SELECT 
    product_id,
    product_name,
    created_at,
    ROW_NUMBER() OVER (PARTITION BY product_id ORDER BY created_at) AS stock_rank
FROM 
    inventory_items
QUALIFY 
    stock_rank = 1;

"""
db.sql(query)
```

```python
# Obtenir le produit le moins coûteux de chaque catégorie
# Cette requête utilise QUALIFY pour obtenir le produit avec le prix de détail le plus bas dans chaque catégorie de produits.

query = """ 
SELECT 
    category,
    name,
    retail_price,
    ROW_NUMBER() OVER (PARTITION BY category ORDER BY retail_price) AS price_rank
FROM 
    products

QUALIFY 
    price_rank = 1;

"""
db.sql(query)
```
