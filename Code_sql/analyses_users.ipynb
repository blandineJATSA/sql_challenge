```python
# import les packages
import zipfile
import csv
import pandas as pd
import duckdb as db
```

```python
# Remplacez 'chemin_vers_fichier_zip' par le chemin de votre fichier zip
chemin_vers_fichier_zip = 'C:/Users/Moi/sql_challenge/database/events.zip'

# Ouvrir le fichier zip
with zipfile.ZipFile(chemin_vers_fichier_zip, 'r') as z:
    # Liste des fichiers contenus dans le zip
    liste_fichiers = z.namelist()
    # On suppose qu'il n'y a qu'un fichier CSV dans le zip
    nom_fichier_csv = liste_fichiers[0]
    # Lire le fichier CSV dans une dataframe
    with z.open(nom_fichier_csv) as f:
        events = pd.read_csv(f, encoding='latin1')

# Afficher les 5 premières lignes de la dataframe
events.head()
```

```python
query =""" 
select id, sequence_number, created_at, event_type, ip_address, cast(session_id as varchar) as session, user_id 
from events
where session ='1d388364-8080-4e09-a632-9121e336a670'
"""
db.sql(query)
```

```python
query =""" 
select user_id, session_id, count() over(partition by session_id, user_id ) as nombre
from events 
order by nombre DESC
"""
db.sql(query)
```
